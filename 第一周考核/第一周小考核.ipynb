```python
import numpy as np                                      
import pandas as pd
pd.set_option('display.max_rows', None)                   
data = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
data.head()
```

**在上面数据的预览中，`User_Score`这列是包含数字的，但一整列都是object，说明混进了其它类型。**  
**下面统计中，tbd表示待定，可以把它替换为NaN**

```python
data['User_Score'].value_counts()[:20]
```

**再具体看看都是什么类型**

```python
temp = data['User_Score'].copy()
temp = temp.to_frame()
temp['type'] = temp['User_Score'].apply(lambda x : type(x))
temp[:200:20]
```

**之后写个函数，将字符串数字转为int或float, tbd替换为空值**

```python
def Convert(x):
    if pd.isna(x):
        return x
    elif x == 'tbd':
        return np.nan
    else:
        return eval(x)
data['User_Score'] = data['User_Score'].apply(Convert)
```

**大部分都是评分数据的列缺失，索性我就把数据分成有缺失值和没有缺失值两部分**  
**data2的评分数据大部分都是空值，于是就把这一部分的列删掉**

```python
data1 = data[np.sum(data.isnull(), axis = 1) == 0].copy()
data2 = data[np.sum(data.isnull(), axis = 1) > 0].copy()
data2.drop(columns = ['Critic_Score', 'Critic_Count', 'User_Score', 'User_Count'], inplace = True)
```

**貌似 `User_Score` 和 `Critic_Score` 看起来有点像，一个十分，一个百分。**  
**emm。。来看看它们的相关性**

```python
cov = np.sum((data1['User_Score'] - data1['User_Score'].mean()) * \
             (data1['Critic_Score'] - data1['Critic_Score'].mean()))
std = np.sqrt(np.sum((data1['User_Score'] - data1['User_Score'].mean())**2) * \
              np.sum((data1['Critic_Score'] - data1['Critic_Score'].mean())**2))
cov / std
```
